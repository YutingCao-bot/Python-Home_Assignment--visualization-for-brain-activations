```python
#load the data
import nibabel as nib
anatomical_image = nib.load('anatomical-MRI data.nii')
functional_image = nib.load('uniformity-functional data social-cognition.nii')
#get functional data(convert the image data into a Numpy arrary using get_fdata)
data = functional_image.get_fdata()







```

This documentation explains the functionality of the second code cell:

I used two packages: nilearn and nibabel. You can find detailed explanations about these packages through the following links:

Nilearn - https://nilearn.github.io/dev/modules/generated/nilearn.plotting.plot_stat_map.html 

Nibabel - https://nipy.org/nibabel/nibabel_images.html

I think the attention should be focused on the plot_stat_map function from the nilearn package.


```python

#visualize my functional data ontop of the anatomical scan
from nilearn import plotting
import nibabel as nib
anatomical_image = nib.load('anatomical-MRI data.nii')
functional_image = nib.load('uniformity-functional data social-cognition.nii')
#visualize image
plotting.plot_stat_map(
    functional_image, 
    bg_img=anatomical_image,
    threshold=3.0, 
    colorbar=True, 
    title='Brain regions typically associated with social cognition', 
    cut_coords=None, 
    display_mode='ortho',
    cmap="hot",
      )
plotting.show()


```

This documentation explains the functionality of the third code cell:

In the final step of the first code cell, I write code to convert the image data into a NumPy array. That’s why I need to import the numpy package before plotting. 

I use the matplotlib.pyplot package to plot the histogram. You can learn more about the histogram function by searching for it on Google.

```python
#plot a histogram of the values in the functional data file
import numpy as np
import matplotlib.pyplot as plt
flattened_data = data.flatten()
#remove the background values(keep the positive values)
positive_data = flattened_data[flattened_data > 0]
#plotting
plt.hist(positive_data, bins=100, color='green', alpha=0.7)
plt.title('Histogram of Functional Data Values')
plt.xlabel('Intensity Value')
plt.ylabel('Frequency')
plt.show()
```
